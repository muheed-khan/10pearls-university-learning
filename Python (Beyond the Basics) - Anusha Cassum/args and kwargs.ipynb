<h1>args and kwargs</h1>

Packing and unpacking of asterik operators

```python
temp = ['a','b','c'] #packing
a,b,c = temp #unpacking
print(a, b, c)
```

```python
*temp1, = 'a', 'b', 'c' #packing with asterik operator
print(temp1)
```

```python
i , *j = temp1 #unpacking with asterik
print(i,j)
```

The above condition will run when at-least one element is present.

```python
grade = sorted([ 50, 80, 70, 60]) #sorted array of grades, for finding highest and lowest using asterik operator
lowest , *useless, highest = grade

print("Lowest : ",lowest)
print("Useless : ",useless)
print("Highest : ",highest)
```
