<h1>List/Dictionary/set Comprehensions</h1>

List Comprehensions

```python
#normal code
numbers = [i for i in range(10)]
even_num = []
for even_no in numbers:
    if even_no % 2 == 0:
        even_num.append(even_no)
    
print(even_num)


#list comprehension

even_num_comprehension = [even_no_c for even_no_c in numbers if even_no_c % 2 == 0]
print(even_num_comprehension)
```

Dictionary Comprehensions

```python
#normal code
people = {'sameer' : 11, 'muheed' : 15, 'taha' : 16, 'rana' : 13}
peep_adults_after_5years = {}
for key,val in people.items():
    if (val + 5) >= 18:
        peep_adults_after_5years.update({key:val})
print(peep_adults_after_5years)


# Dictionary Comprehension

peep_adults_after_5years_comprehension = {key:val for key,val in people.items() if (val+5) >= 18}
print(peep_adults_after_5years_comprehension)
```

Set Comprehension

```python
#normal Code
num_list = [1,2,3,4,4,5,6,6,7,8,9,9]
non_duplicate_set_of_even = set()

for num in num_list:
    if num % 2 == 0:
        non_duplicate_set_of_even.add(num)

print(non_duplicate_set_of_even)



#set comprehension
non_duplicate_set_of_even_comprehension = set()
non_duplicate_set_of_even_comprehension = {num for num in num_list if num % 2 == 0}
print(non_duplicate_set_of_even_comprehension)
```
